```python
import pandas as pd
import time
import requests
from bs4 import BeautifulSoup

#Load file
path = '2017_18CMS'
CMS_data = pd.read_excel(path, sheet_name = '2017_18CMS')

code_list = []
translation = []
for index, row in CMS_data.iterrows():
    for di in range(1,26):
        diga_no = 'DIAGNOSIS_CODE_'+str(di)
        # Make a request to the website
        scode = str(row[diga_no])
        if scode not in code_list:
            code_list.append(scode)
            slink = 'https://www.icd10data.com/search?s='
            ssearch = slink + scode
            r = requests.get(ssearch)
            # Create an object to parse the HTML format
            soup = BeautifulSoup(r.content, 'html.parser')
            # Retrieve all popular news links (Fig. 1)
            cormod = soup.find_all('div', {'class':'searchPadded'})
            for points in cormod:
                point = str(points.text)
            translation.append(point)
            time.sleep(5)
        #print(scode)
        #print(point)
new_list = {'code':code_list, 'translation': translation}
df = pd.DataFrame(new_list)
writer = pd.ExcelWriter('code_meaning.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='welcome', index=False)
writer.save()
```
